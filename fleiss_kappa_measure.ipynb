```python
columns = ['Male','Female','Unsure']
index = []
f_count = 0
m_count = 0
u_count = 0

f_list = []
m_list = []
u_list = []



for column in survey_data:
    #print(survey_data[column].value_counts())
    index.append(column)

data = pd.DataFrame(index = index,columns = columns)

for column in survey_data:

    f_count = 0
    m_count = 0
    u_count = 0
    #print(column)
    
    for row in survey_data[column]:



        if row == 'Male':
            m_count = m_count + 1

        elif row == 'Female':
            f_count = f_count + 1

        else:
            u_count = u_count + 1

    #print(m_count,f_count,u_count)
    m_list.append(m_count)
    f_list.append(f_count)
    u_list.append(u_count)
    
    

data['Male'] = m_list
data['Female'] = f_list
data['Unsure'] = u_list
```

```python
raters = 6
categories = 3
subjects = 500

sum_cells = subjects * raters

total_m = data['Male'].sum()
total_f = data['Female'].sum()
total_u = data['Unsure'].sum()

#print(total_m, total_f, total_u)

#proportion of assignments that were made to the category of male, female and unsure
assign_male = round((total_m/sum_cells),3)
assign_female = round((total_f/sum_cells),3)
assign_unsure = round((total_u/sum_cells),3)

#print(assign_male,assign_female,assign_unsure)

mean_assign =  round((assign_male**2) + (assign_female**2) + (assign_unsure**2),3)

#print(mean_assign)

# degree of agreement that is attainable above chance
deg_agreement = 1- mean_assign
print(deg_agreement)
```

```python
# calculating the extent to which the raters agree

def cal_prob(row):
    x = ( (row['Male'] ** 2) + (row['Female'] ** 2) + (row['Unsure'] ** 2) - raters)
    prob = x/( raters * (raters - 1) )
    return round(prob,3)

data.apply(cal_prob , axis = 1)
```

```python
data['Probability'] = data.apply(cal_prob, axis =1)
data
```

```python
# calculate the mean aggrement for measuring kappa

prob_mean = round(data['Probability'].sum() / subjects , 3)

# degree of agreement actually achieved above chance (prob_mean - mean_assign )
agree_abv_chance = round( (prob_mean - mean_assign) , 3)

print(agree_abv_chance)
```

```python
# kappa measure

kappa_measure = round( (agree_abv_chance / deg_agreement) , 3)
print(kappa_measure)
```
