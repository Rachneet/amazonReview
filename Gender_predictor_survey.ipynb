```python
#read male data
amazon_df_m = pd.read_csv('/media/backup/Data/Amazon/amazon_male.csv', sep='|',encoding='utf8',quoting=csv.QUOTE_NONE, dtype={'Overall_Rating':float, 'Timestamp': int})
```

```python
#unique entries
unique_males = amazon_df_m['Name'].unique().tolist()
```

```python
#read female data
amazon_df_f = pd.read_csv('/media/backup/Data/Amazon/amazon_female.csv', sep='|',encoding='utf8',quoting=csv.QUOTE_NONE, dtype={'Overall_Rating':float, 'Timestamp': int})
```

```python
#unique entries
unique_females = amazon_df_f['Name'].unique().tolist()
```

```python
names_df = pd.read_excel("/home/rachneet/Desktop/name_list.xls")
```

```python
names_list = []

for row in names_df['Names']:
    names_list.append(row)
    
```

```python
len(names_list)
```

```python
#creating dict for males
males_dict ={}

for name in names_list:
    if name in unique_males:
        males_dict[name] = 'Male'
        
males_dict
        
```

```python
#creating dict for females
females_dict ={}

for name in names_list:
    if name in unique_females:
        females_dict[name] = 'Female'
        
females_dict
```

```python
def merge_two_dicts(x, y):
    z = x.copy()   # start with x's keys and values
    z.update(y)    # modifies z with y's keys and values & returns None
    return z
```

```python
guesser_dict = merge_two_dicts(males_dict, females_dict)
```

```python
#comparing the 2 dictionaries

shared_items = {k: result_dict[k] for k in result_dict if k in guesser_dict and result_dict[k] == guesser_dict[k]}
print(len(shared_items))
```

```python
#calculating accuracy of gender guesser

accuracy = (len(shared_items)/len(guesser_dict))*100
print(accuracy)
```

```python
#names that could not be identified

undecided = []

for k in guesser_dict:
    if k not in shared_items:
        undecided.append(k)
        
undecided
```
